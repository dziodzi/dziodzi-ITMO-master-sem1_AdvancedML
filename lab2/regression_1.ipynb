```python
df = pd.read_csv("/content/drive/MyDrive/DeepLearningCourse/task2_1.csv")
df
```

```python
sns.scatterplot(x=df.X, y=df.Y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(df[['X']], df[['Y']])
```

```python
print(f'Значение X: {round(df.X.mean(), 3)} \nЗначение Y: {round(df.Y.mean(), 3)}')
```

```python
print(f'Значение theta_0: {round(reg.intercept_.mean(), 3)} \nЗначение theta_1: {round(reg.coef_.mean(), 3)}')
```

```python
theta1 = sum((df.X - df.X.mean()) * (df.Y - df.Y.mean()))/sum((df.X - df.X.mean()) ** 2)
theta0 = df.Y.mean() - theta1 * df.X.mean()
print(f'Значение theta_0: {round(theta0, 3)} \nЗначение theta_1: {round(theta1, 3)}')
```

```python
x = np.linspace(df.X.min(), df.X.max(), 100)
y = theta0 + theta1 * x
sns.scatterplot(x=df.X, y=df.Y)
sns.lineplot(x=x, y=y, color='red')
```

```python
R2 = 1 - sum((df.Y - theta0 - theta1 * df.X) ** 2)/sum((df.Y - df.Y.mean()) ** 2)
print(f'Значение R2: {round(R2, 3)}')
```
